## Overview About the Problem

### What is Food Security?
According to the United Nations' (UN) Committee on World Food Security, food security is defined as people, "at all times", having "Physical, Social, and Economic" access to "sufficient, safe and nutritious food" that meets their "dietary needs for an active and healthy life" (FAO, 1996).

#### Personas
- Persona 1: A food security analyst working for a government agency. He/She helps to inform policymakers and requires a model to predict the level of food security of the country in the next few years, given the current data.    

- Persona 2: A member of international organisations such as the Food and Agriculture Organisation. He/She aims to “improve nutrition, increase agricultural productivity, raise the standard of living in rural populations” on a global scale, and they need a model to predict the data for each factor such that food security is achieved, or a model to measure how long it might take for a country to attain a desirable level of food security.   

- Persona 3: A decision-maker in a Non-Profit Organisation, which focuses their efforts on countries that require the most aid. He/She requires a model to identify countries of concern in both present and future.


Keeping the needs of our personas in mind, we aim to come up with a model that helps them to predict the level of food security of a country.

### Problem Statement
In line with the definition of food security provided by UN, we have decided to find out how physical, social, and economic factors have affected the dietary requirements for an active and healthy lifestyle of the population of each country. As a result, we have come up with the following problem statement: **“How might we predict whether the dietary needs for individuals are met for each country, considering a country’s economy, and its citizen’s social and physical environment?”**

#### Economic Factors
We define _economic factors_ as factors that is due to the state of the economy around the world, uncontrollable by individuals.
- Percentage of population employed in agriculture, forestry, and fishing in percent:      
We believe that the greater the percentage of population allocated to agriculture, the higher the production of crops. This allows greater harvest and higher self-sufficiency on local supply of food which results in a greater access to food domestically (Bishwajit, G., 2013).
- Gross Domestic Product (GDP) per capita, adjusted for PPP in USD per capita:     
An increase in GDP per capita leads to an increase in purchasing power of individuals. This makes food more affordable to the individual and increasing his access to food (Fernandes, M. & Samputra, P., 2022). Hence, we will focus on GDP per capita as our economic factor and see how this affects food security in general.

#### Physical Factors
We define _physical factors_ as individuals having a small portion of control over these factors; of which we have two factors.
- Percentage of population having eating disorders in percent:   
There are four types of eating disorders, of which some result in overconsumption while others lead to underconsumption of food (AwareSG, n.d.). However, our main concern here is that food security not just about the physical means of being able to consume food but the mental means as well. Eating disorders can affect individuals having a mental "insecurity" with food which results in an unhealthy lifestyle.
- Land used for agriculture per capita in km2 per capita:   
A greater agricultural land per capita ratio suggests that more land is allocated to grow crops per person. “Agriculture provides the largest share of food supplies” (M. Viana, C., 2021), thus it is safe to say that the greater land per capita, the greater the increase in agriculture food production and security allowing more access to sufficient food.

#### Social Factors
We define _social factors_ as individuals not being able to control these factors; of which we have two factors.
- Percentage of population having access to basic drinking water services in percent:      
Water is essential to food security (United Nations, n.d.). Water is used in many ways to ensure food security; from growing crops, to cleaning and cooking food. Hence, we have also considered the relationship between access to basic drink water relates and access to sufficient, safe, and nutritious food.
- Carbon Dioxide (CO2) emitted for each country per agricultural land area in Million Metric Tons (MMT)/km2:   
CO2 increases rate of photosynthesis hence increasing crop yields. CO2 also allows plants to “accumulate more carbohydrates” (L. Ebi, K., 2019), increasing calorie yield, which aligns with our model since we are focusing on calories per capita per day. Considering the nutritional requirements of individuals, this factor gives us insight on how the access to sufficient and nutritious food increases food security.

#### Dependent Value
Our dependent variable will be the ratio of average calories supplied in kcal per capita per day to average calorie requirements per capita per day in kcal per capita per day, known as Y ratio.
To measure food security, more specifically the dietary needs for an active and healthy life, we will be calculating it using the ratio of average calories supplied per individual per day over the average calorie requirement per individual per day. A higher ratio is indicative of greater food security, and vice versa. However, we note that the average calories supplied is not indicative of the calories that are consumed and realistically shows the calories that are available for consumption on a retail level (Roser, M., 2013). 
$$
 \text{Y Ratio} = \frac{\text{Daily calories supplied (kcal/capita day)}}{\text{Minimum calories required (kcal/capita day)}}
$$

#### Removal of outliers

```python
# write function to return the upper limit and lower limit based on IQR
# df = dataframe; i = the name of the column
def identify_upper_lower (df, i):
    q3 = df[i].quantile (q = 0.75)
    q1 = df[i].quantile (q = 0.25)
    IQR = q3 - q1
    upper_limit = q3 + 1.5 * IQR
    lower_limit = q1 - 1.5 * IQR
    return upper_limit, lower_limit
```

### Training the $1^{st}$ Model

Defining functions used in class

```python
def get_features_targets(df, feature_names, target_names):
    df_feature = df.loc[:,feature_names]
    df_target = df.loc[:,target_names]
    return df_feature, df_target

def prepare_feature(df_feature):
    feature = df_feature.to_numpy().reshape(-1, len(df_feature.columns))
    X = np.concatenate((np.ones((feature.shape[0], 1)),feature), axis = 1)
    return X

def prepare_target(df_target):
    target = df_target.to_numpy().reshape(-1, len(df_target.columns))
    return target

def predict(df_feature, beta):
    X = prepare_feature(df_feature)
    return np.matmul(X, beta)

def calc_linear(X, beta):
    return np.matmul(X, beta)

def split_data(df_feature, df_target, random_state=None, test_size=0.5):
    indexes = df_feature.index
    if random_state != None:
        np.random.seed(random_state)
    
    k = int(test_size*len(indexes))
    test_index = np.random.choice(indexes, k, replace = False)
    indexes = set(indexes)
    test_index = set(test_index)
    train_index = indexes - test_index
    
    df_feature_train = df_feature.loc[train_index,:]
    df_feature_test = df_feature.loc[test_index,:]
    
    df_target_train = df_target.loc[train_index,:]
    df_target_test = df_target.loc[test_index,:]
    return df_feature_train, df_feature_test, df_target_train, df_target_test

def r2_score(y, ypred):
    ymean = np.mean(y)
    diff = y - ymean
    sstot = np.matmul(diff.T, diff)
    error = y - ypred
    ssres = np.matmul(error.T,error)
    return 1 - ssres/sstot

def mean_squared_error(target, pred):
    n = target.shape[0]
    error = target - pred
    return (1/n)*np.matmul(error.T, error)[0][0]

def compute_cost(X, y, beta):
    J = 0
    m = X.shape[0]
    error = calc_linear(X,beta)-y
    error_sq = np.matmul(error.T,error)
    J = (1/(2*m)) * error_sq
    return J

def gradient_descent(X, y, beta, alpha, num_iters):
    m = X.shape[0]
    J_storage = np.zeros((num_iters,1))
    for n in range(num_iters):
        deriv = np.matmul(X.T,calc_linear(X,beta)-y)
        beta = beta - (alpha * (1/m)) * deriv
        J_storage[n] = compute_cost(X,y,beta)
    return beta, J_storage
```

### Considering Possible Removal of Features for $1^{st}$ model

- We will be using another metric to verify whether there is a good correlation between between our features and the target. We decided to use Mean Absolute Error (MAE).
$$\large \textrm{MAE} = \frac{1}{n} \sum |y_i - x_i| %$$

```python
def mean_absolute_error(df,feature,target):
    x = df[feature].to_numpy()
    y = df[target].to_numpy()
    n = df.shape[0]
    error = np.abs((y-x))
    return (1/n) * (error.sum())
```

```python
def mean_absolute_percentage_error(df, feature, target):
    x = df[feature].to_numpy()
    y = df[target].to_numpy()
    n = df.shape[0]
    error = np.abs((y-x)/y)
    return ((1/n) * (error.sum())) / 100
```

#### Calculate Y ratio 
$$
\frac{\text{Daily calories supplied (kcal/capita day)}}{\text{Minimum calories required (kcal/capita day)}}
$$

```python
# extract values from countries in combined countries list
df_calories_cc3 = df_calories[df_calories["Country"].isin(cc3)]
df_min_cal_cc3 = df_min_cal[df_min_cal["Country"].isin(cc3)]

df_y_ratio_3 = df_calories_cc3.copy()
# extract minimum calorie value
df_y_ratio_3["Mininum_calorie_intake"] = list(df_min_cal_cc3["Mininum_calorie_intake"])
# calculate Y ratio based on stated equation
df_y_ratio_3["y_ratio"] = df_y_ratio_3["Calories_supply"]/df_y_ratio_3["Mininum_calorie_intake"]
# set index from 0
df_y_ratio_3 = df_y_ratio_3.reset_index(drop=True)
# print dataframe
## df_y_ratio_3
```

#### Calculate Carbohydrate supplied (g/capita day)

We can easily calculate the carbohydrate supplied using
$$
\small \text{Carbohydate Supplied} = \frac{\text{Energy supplied - (Protein supplied * 4) - (Fat supplied * 9 )}}{\text{4}}
$$
- 1g of Carbohydrate = 4 kcal
- 1g of Protein = 4 kcal
- 1g of Fat = 9 kcal

```python
# extract values from countries in combined countries list
df_protein_cc3 = df_protein[df_protein["Country"].isin(cc3)]
df_fat_cc3 = df_fat[df_fat["Country"].isin(cc3)]
df_calories_cc3 = df_calories[df_calories["Country"].isin(cc3)]

df_carb_cc3 = df_protein_cc3.copy()
# extract Fat value
df_carb_cc3["Fat"] = list(df_fat_cc3["Fat"])
# extract Calories supply value
df_carb_cc3["Total_calories"] = list(df_calories_cc3["Calories_supply"])
# calculate carbohydrate in (g/capita day)
df_carb_cc3["Carbohydrate"] = (df_carb_cc3["Total_calories"] - 9*df_carb_cc3["Fat"] - 4*df_carb_cc3["Protein"])/4
# set index from 0
df_carb_cc3 = df_carb_cc3.reset_index(drop=True)
# print dataframe
## df_carb_cc3
```
